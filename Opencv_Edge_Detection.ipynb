```python
#import library
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read img
bird_img = cv2.imread("gary-bendig-WPmPsdX2ySw-unsplash.jpg")
blured_img = cv2.blur(bird_img, ksize = (5,5))
```

```python
plt.imshow(bird_img)
```

```python
#calc threshold with median
median = np.median(bird_img)
lower = int(max(0, 0.3 * median))
upper = int(min(255, 1.7 * median))
```

```python
#detect edges with Canny Edge Detection
edg = cv2.Canny(blured_img, lower/10, upper/6)
plt.imshow(edg, )
```
